```python
import numpy as np
```

```python
deliveries = np.genfromtxt(
    'deliveries.csv',
    delimiter=',',
    names=True,
    dtype=None,
    encoding='utf-8'
)
print(deliveries)
```

Required columns

```python
match_id = deliveries['match_id']
batting_team = deliveries['batting_team']
batter = deliveries['batter']
bowler = deliveries['bowler']
batsman_runs = deliveries['batsman_runs']
over = deliveries['over']
ball = deliveries['ball']
```

Task 1: Total Runs per Match

```python
total_runs = deliveries['total_runs']

unique_ids , inverse_index = np.unique(match_id, return_inverse=True)
runs = np.bincount(inverse_index, weights = total_runs)
res = list(zip(unique_ids.tolist(), runs.tolist()))
print(res)
```

Task 2: Top 5 Batters

```python
unique_batters, inverse_batter = np.unique(batter, return_inverse=True)
batter_runs = np.bincount(inverse_batter, weights = batsman_runs)
top_indices = np.argsort(batter_runs)[-5:][::-1]
top_batters = [(unique_batters[i], batter_runs[i]) for i in top_indices]
print(top_batters)
```

Task 3: Strike Rate of Batters

```python
# Strike Rate = (Total Runs / Balls Faced) × 100
# min 100 balls faced

batter_balls = np.bincount(inverse_batter)
strike_rates = (batter_runs / batter_balls) * 100
qualified_indices = np.where(batter_balls >= 100)[0]
qualified_strike_rates = [(unique_batters[i], strike_rates[i]) for i in qualified_indices]
qualified_strike_rates.sort(key=lambda x: x[1], reverse=True)
print(qualified_strike_rates)
```

Task 4: Economy Rate of Bowlers

```python
#Economy = Runs Conceded / Overs Bowled
#Convert balls to overs (balls ÷ 6)
# min 30 overs bowled
unique_bowlers, inverse_bowler = np.unique(bowler, return_inverse=True)
bowler_balls = np.bincount(inverse_bowler)
bowler_runs = np.bincount(inverse_bowler, weights=total_runs)
bowler_overs = bowler_balls / 6
economy = bowler_runs / bowler_overs
qualified_indices = np.where(bowler_balls >= 300)[0]
sorted_indices = qualified_indices[np.argsort(economy[qualified_indices])]
bowler_economy = [(unique_bowlers[i], economy[i]) for i in sorted_indices]
print(bowler_economy)
```

Task 5: Runs per Over

```python
runs_per_over = np.bincount(over, weights=total_runs, minlength=20)
balls_per_over = np.bincount(over, minlength=20)
avg_runs_per_over = (runs_per_over / balls_per_over)*6
print(avg_runs_per_over)
```

Task 7: Death Overs Analysis


```python
death_mask = (deliveries['over'] >= 15)
total_death_runs = np.sum(deliveries['total_runs'][death_mask])
print("Total runs scored in the death overs: " + str(total_death_runs))
```

```python
import numpy as np

# 1. Death overs mask (16–20)
death_mask = (deliveries['over'] >= 16) & (deliveries['over'] <= 20)

# 2. Normalize team names
name_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Rising Pune Supergiant': 'Rising Pune Supergiants'
}

death_teams = np.array([
    name_map.get(t, t) for t in deliveries['batting_team'][death_mask]
])

# 3. Runs in death overs
death_runs = deliveries['total_runs'][death_mask]

# 4. Group by team
unique_teams, inverse = np.unique(death_teams, return_inverse=True)
team_runs = np.bincount(inverse, weights=death_runs)

# 5. Get top team
top_idx = np.argmax(team_runs)

top_team = unique_teams[top_idx]
top_runs = team_runs[top_idx]

print(top_team, int(top_runs))
```

Task 8: Highest Scoring Match

```python
unique_matches , inverse = np.unique(deliveries['match_id'], return_inverse = True)
match_runs = np.bincount(inverse, weights=deliveries['total_runs'])

top_index = np.argmax(match_runs)
top_match_id = unique_matches[top_index]
top_runs = match_runs[top_index]
print((top_match_id, top_runs))
```

Task 9: Runs per Team per Match


```python
pairs = list(zip(deliveries['match_id'], deliveries['batting_team']))
unique_pairs, inverse = np.unique(pairs, axis=0, return_inverse=True)
pair_runs = np.bincount(inverse, weights=deliveries['total_runs'])

result = [
    (unique_pairs[i][0], unique_pairs[i][1], pair_runs[i])
    for i in range(len(unique_pairs))
]

print(result)
```

Task 10: Match Winner Approximation

```python
match_winners = []

for i in unique_ids:
    match_mask = match_id == i
    teams = np.unique(batting_team[match_mask])
    
    if len(teams) != 2:
        continue
    
    team1, team2 = teams
    team1_runs = np.sum(total_runs[match_mask & (batting_team == team1)])
    team2_runs = np.sum(total_runs[match_mask & (batting_team == team2)])
    
    if team1_runs > team2_runs:
        winner = team1
    elif team2_runs > team1_runs:
        winner = team2
    else:
        winner = 'Tie'
        
    match_winners.append((i, team1, team1_runs, team2, team2_runs, winner))
print(match_winners)
```

Task 11: Match Scorecard Generation


```python
# 1. Combine match_id + team
pairs = np.core.records.fromarrays(
    [deliveries['match_id'], deliveries['batting_team']],
    names='match_id,team'
)

# 2. Unique pairs + mapping
unique_pairs, inverse = np.unique(pairs, return_inverse=True)

# 3. Runs per (match, team)
pair_runs = np.bincount(inverse, weights=deliveries['total_runs'])

# 4. Convert to structured list
data = [
    (unique_pairs[i][0], unique_pairs[i][1], pair_runs[i])
    for i in range(len(unique_pairs))
]

# 5. Group by match_id
from collections import defaultdict
match_dict = defaultdict(list)

for match_id, team, runs in data:
    match_dict[match_id].append((team, runs))

# 6. Print scorecard
for i, (match_id, teams) in enumerate(match_dict.items(), start=1):
    print(f"Match {i}:")
    for team, runs in teams:
        print(f" {team}: {int(runs)} runs")
    print()
```
