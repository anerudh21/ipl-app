## Stage 2: Data Cleaning & Validation

```python
#finding null values
missing = deliveries_df.isnull().sum()
print(missing)
```

```python
deliveries_df['extras_type'] = deliveries_df['extras_type'].fillna('normal')
```

```python
new_missing = deliveries_df.isnull().sum()
print(new_missing)
```

```python
match_missing = matches_df.isnull().sum()
print(match_missing)
```

Updating empty cities to Dubai (checked in the excel) and filling all null result margin values with 0 (because game was tied)

```python
matches_df['city'] = matches_df['city'].fillna('Dubai')
print(matches_df['city'].isnull().sum())
```

```python
matches_df['result_margin'] = matches_df['result_margin'].fillna(0)
print(matches_df['result_margin'].isnull().sum())
```

```python
#show null values rows for winner
matches_df[matches_df['winner'].isnull()]
```

```python
#drop duplicate rows if any
matches_df = matches_df.drop_duplicates()
```

Stage 3: Data Transformation

```python
#total runs per ball 
deliveries_df['calculated_total_runs'] = deliveries_df['batsman_runs'] + deliveries_df['extra_runs']

#is_boundary' flag 
deliveries_df['is_boundary'] = deliveries_df['batsman_runs'].isin([4, 6])
```

```python
name_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Rising Pune Supergiant': 'Rising Pune Supergiants'
}

deliveries_df['batting_team'] = deliveries_df['batting_team'].replace(name_map)
deliveries_df['bowling_team'] = deliveries_df['bowling_team'].replace(name_map)
matches_df['team1'] = matches_df['team1'].replace(name_map)
matches_df['team2'] = matches_df['team2'].replace(name_map)
```

```python
# Merge the two datasets
unified_df = deliveries_df.merge(matches_df, left_on="match_id", right_on="id", how="inner")

#drop the redundant 'id' column from the matches_df side
unified_df.drop('id', axis=1, inplace=True)
```

```python
# Check shape: Rows should equal deliveries_df rows, columns will be sum of both
print(f"Unified DataFrame Shape: {unified_df.shape}")
```

```python
unified_df.head()
```

## Stage 4: Core Analysis

```python
# 1. Total Runs per Match
total_runs_match = unified_df.groupby('match_id')['total_runs'].sum()
print(total_runs_match)
```

```python
# 2. Runs per Team per Match
runs_team_match = unified_df.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()
print(runs_team_match)
```

```python
# 3 & 4. Top 10 Batters and Strike Rate
batter_stats = unified_df.groupby('batter').agg({'batsman_runs': 'sum', 'ball': 'count'})
batter_stats.rename(columns={'batsman_runs': 'runs', 'ball': 'balls_faced'}, inplace=True)

# Calculate Strike Rate: (Runs / Balls) * 100
batter_stats['strike_rate'] = (batter_stats['runs'] / batter_stats['balls_faced']) * 100

top_10_batters = batter_stats.sort_values(by='runs', ascending=False).head(10)

print(top_10_batters)
```

```python
# Economy = (Total Runs Conceded / Total Overs Bowled)
bowler_stats = unified_df.groupby('bowler').agg({'total_runs': 'sum', 'ball': 'count'})
bowler_stats['overs'] = bowler_stats['ball'] / 6
bowler_stats['economy'] = bowler_stats['total_runs'] / bowler_stats['overs']

# Filter for bowlers who have bowled at least 10 overs to avoid outliers
top_10_bowlers = bowler_stats[bowler_stats['overs'] >= 10].sort_values(by='economy').head(10)

print(top_10_bowlers)
```

```python
# 7. Highest Individual Score
high_scores = unified_df.groupby(['match_id', 'batter'])['batsman_runs'].sum().reset_index()
max_score = high_scores.sort_values(by='batsman_runs', ascending=False).head(1)
print(max_score)
```

```python
# 8 & 9. Boundary Analysis
unified_df['is_four'] = (unified_df['batsman_runs'] == 4).astype(int)
unified_df['is_six'] = (unified_df['batsman_runs'] == 6).astype(int)

boundary_stats = unified_df.groupby('batter').agg({'batsman_runs': 'sum', 'is_four': 'sum', 'is_six': 'sum'})
boundary_stats['boundary_runs'] = (boundary_stats['is_four'] * 4) + (boundary_stats['is_six'] * 6)
boundary_stats['boundary_pct'] = (boundary_stats['boundary_runs'] / boundary_stats['batsman_runs']) * 100
print(boundary_stats)
```

```python
# 10. Dot Ball Analysis
dot_balls = unified_df[unified_df['total_runs'] == 0].groupby('bowler').size().sort_values(ascending=False)
print(dot_balls)
```

```python
# 12. Powerplay Performance (Overs 1-6)
powerplay_df = unified_df[unified_df['over'] < 6]
powerplay_runs = powerplay_df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
print(powerplay_runs)
```

```python
# Toss impact analysis
matches_df['toss_winner_won'] = matches_df['toss_winner'] == matches_df['winner']
toss_impact = matches_df['toss_winner_won'].value_counts(normalize=True) * 100
print(f"Percentage of teams winning toss and match: {toss_impact[True]:.2f}%")
```

```python
# 17. Venue-wise Analysis
venue_stats = matches_df.groupby('venue').agg({'id': 'count', 'target_runs': 'mean'})
venue_stats.rename(columns={'id': 'total_matches', 'target_runs': 'avg_score'}, inplace=True)
print(venue_stats)
```

```python
# 19. Season-wise Run Trends
season_runs = unified_df.groupby('season')['total_runs'].sum()
print(season_runs)
```

## Stage 5: Derived Insights

1. Most Consistent Batter
2. Best death-over team
3. High-scoring venues

```python
# Calculate matches played per batter
matches_per_batter = unified_df.groupby('batter')['match_id'].nunique()

# Calculate total runs and average
batter_consistency = unified_df.groupby('batter').agg({'batsman_runs': 'sum'})
batter_consistency['matches'] = matches_per_batter
batter_consistency['avg_per_match'] = batter_consistency['batsman_runs'] / batter_consistency['matches']

# Filter for batters who played at least 2 full seasons (approx 30 matches)
most_consistent = batter_consistency[batter_consistency['matches'] >= 30].sort_values(by='avg_per_match', ascending=False).head(5)
print(most_consistent)
```

```python
# Filter for overs 16 to 20
death_overs_df = unified_df[unified_df['over'] >= 16]

# Calculate runs and balls per team
death_stats = death_overs_df.groupby('batting_team').agg({'total_runs': 'sum', 'ball': 'count'})
death_stats['run_rate'] = (death_stats['total_runs'] / death_stats['ball']) * 6

best_death_team = death_stats.sort_values(by='run_rate', ascending=False)
print(best_death_team)
```

```python
# Group by venue and calculate average total runs per match
# We use the match_df here for cleaner match-level totals
venue_scoring = matches_df.groupby('venue')['target_runs'].mean().sort_values(ascending=False)

# Top 5 high-scoring grounds
top_venues = venue_scoring.head(5)
print(top_venues)
```
